## Tiny Project: heart disease prediction

We worked on the heart disease dataset during the DS/ML class. Try to find a model from HF that can predic the diseased cases. You can try multiple models to find the one with the highest accuracy.

We want to automatically classify whether a new case is healthy or has a heart disease.

Just like the text classification, instead of training a model ourselves, we will reuse a pretrained model from Hugging Face.
